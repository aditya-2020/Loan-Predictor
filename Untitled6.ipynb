```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
payment_history_data= pd.read_csv(r'D:\cassandra\payment_history_data.csv')
train_data= pd.read_csv(r'D:\cassandra\train_data.csv')
print(payment_history_data.shape)
print(train_data.shape)
payment_history_data.head()
```

```python
payment_history_data.drop('p_limit',axis=1,inplace=True) #remove column p_limit
print(payment_history_data.shape)
print(train_data.shape)
payment_history_data.head()
```

```python
payment_history_data.drop('recent_payment_activity',axis=1,inplace=True) #remove col recent_pay..
print(payment_history_data.shape)
print(train_data.shape)
payment_history_data.head()
```

```python
payment_history_data = payment_history_data.dropna(how='any',axis=0)  #remove rows with null values 

#mean=train_data['Beta'].mean() #calculate mean of Beta column
train_data['Beta'].fillna(value=train_data['Beta'].mean(), inplace=True)

train_data = train_data.dropna(how='any',axis=0)

print(payment_history_data.shape)
print(train_data.shape)
payment_history_data.head()
```

```python
grouped=payment_history_data.groupby('id').agg([np.average]) #group is data that we need, take average of same id 
print(payment_history_data.shape) 
print(train_data.shape)
payment_history_data.head()
```

```python
grouped.head()
```

```python
grouped.shape
```

```python
grouped_copy=grouped.copy() #copying grouped just in case
```

```python
yplusx = pd.merge(grouped, train_data, on='id', how='inner') #merge both data to form final dataset
yplusx.drop('id',axis=1,inplace=True)
```

```python
yplusx.head()
```

```python
yplusx.shape
```

```python
yplusx=yplusx.sample(frac=1) #shuffle the entire data
```

```python
val=yplusx[0:50] #split data into validation and train set
train=yplusx[50:]
```

```python
val.head()
```

```python
print(val.shape)
print(train.shape)
```

```python
y_val=val['label']
x_val=val.loc[:, val.columns != 'label']
y_train=train['label']
x_train=train.loc[:, train.columns != 'label']
```

```python
print(y_val.shape)
print(x_val.shape)
print(y_train.shape)
print(x_train.shape)
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

x_train=normalize(x_train)
x_val=normalize(x_val)
```

```python
x_train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
```

```python
predictions = logmodel.predict(x_val)
print(classification_report(y_val, predictions))
print(confusion_matrix(y_val, predictions))
print(accuracy_score(y_val, predictions))
```
