```python
# PLEASE NOTE: we have added comments where ever possible also we have included a short summary of how we found out our final
# model later on.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
payment_history_data= pd.read_csv(r'D:\cassandra\payment_history_data.csv') #read data
train_data= pd.read_csv(r'D:\cassandra\train_data.csv')
print(payment_history_data.shape)
print(train_data.shape)
payment_history_data.head()
```

```python
payment_history_data.drop('p_limit',axis=1,inplace=True) #remove column p_limit
print(payment_history_data.shape)
print(train_data.shape)
payment_history_data.head()
```

```python
payment_history_data.drop('recent_payment_activity',axis=1,inplace=True) #remove col recent_pay..
print(payment_history_data.shape)
print(train_data.shape)
payment_history_data.head()
```

```python
payment_history_data = payment_history_data.dropna(how='any',axis=0)  #remove rows with null values 

#mean=train_data['Beta'].mean() #calculate mean of Beta column
train_data['Beta'].fillna(value=train_data['Beta'].mean(), inplace=True)

train_data = train_data.dropna(how='any',axis=0)

print(payment_history_data.shape)
print(train_data.shape)
payment_history_data.head()
```

```python
grouped=payment_history_data.groupby('id').agg([np.average]) #group is data that we need, take average of same id 
print(payment_history_data.shape) 
print(train_data.shape)
payment_history_data.head()
```

```python
yplusx = pd.merge(grouped, train_data, on='id', how='inner') #merge both data to form final dataset
yplusx.drop('id',axis=1,inplace=True)
```

```python
yplusx.head()
```

```python
yplusx.shape
```

```python
yplusx=yplusx.sample(frac=1) #shuffle the entire data
yplusx.shape
```

```python
val=yplusx[0:200] #split data into validation and train set
train=yplusx[200:]
print(val.shape)
print(train.shape)
```

```python
val.head()
```

```python
print(val.shape)
print(train.shape)
```

```python
y_val=val['label']
x_val=val.loc[:, val.columns != 'label']
y_train=train['label']
x_train=train.loc[:, train.columns != 'label']
```

```python
print(y_val.shape)
print(x_val.shape)
print(y_train.shape)
print(x_train.shape)
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

x_train=normalize(x_train)
x_val=normalize(x_val)
```

```python
x_val.head()
```

```python

from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.feature_selection import mutual_info_regression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.datasets import make_regression

# feature selection
def select_features(x_train, y_train, x_val, i):
    # configure to select a subset of features
    fs = SelectKBest(score_func=mutual_info_regression, k=i)
    # learn relationship from training data
    fs.fit(x_train, y_train)
    # transform train input data
    x_train_fs = fs.transform(x_train)
    # transform test input data
    x_val_fs = fs.transform(x_val)
    return x_train_fs, x_val_fs, fs

for i in range(1,22) :
    x_train_fs, x_val_fs, fs = select_features(x_train, y_train, x_val, i)

# logmodel = LogisticRegression()
# logmodel.fit(x_train_fs, y_train)
#     logmodel = SVC(kernel='linear', 
#     class_weight='balanced', # penalize
#     probability=True)
    logmodel = RandomForestClassifier()
 
    logmodel.fit(x_train_fs, y_train)

    #for i in range(len(fs.scores_)):
     #   print('Feature %d: %f' % (i, fs.scores_[i]))
     # plot the scores
    #pyplot.bar([i for i in range(len(fs.scores_))], fs.scores_)
    #pyplot.show()

    predictions = logmodel.predict(x_val_fs)
    print( roc_auc_score(y_val, predictions) )
    print(classification_report(y_val, predictions))
    print(confusion_matrix(y_val, predictions))
    print(accuracy_score(y_val, predictions))
    
```

```python

# HOW WE FOUND OUR MODEL
# We calculated accuracy for all the values of k (k is the number of features that we select to form the model). And for
# selecting most promising features we use score functions, mutual_info_regression and f_regression. Also we tested our model
# with two algorithms, SVM and Random forest classifier. So in total we had 2*2*21 cases, all of the cases where calculated
# for accuracy and the model with highest F1 score had its result calculated and submitted on kaggle.
# We are writing this because we don't quite remember which setting gave our final score on Kaggle.
```
